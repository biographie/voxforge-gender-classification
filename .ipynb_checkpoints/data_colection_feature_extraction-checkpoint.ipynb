# Classifying gender via audio recordings

## The required data, size and type
In order to be able to classify gender based on audio we need a labelled data set that has the gender label of each voice recording. Voxforge offers such a data set. In it there are recordings of 6247 different people who have recording a number of samples of varying length. The first step will be to scrape the data off of Vox forge

## 1 Scraping the Voxforge website and downloading the data
In this step we will utilize BeautifulSoup to parse the HTML of the website and then extract all the links and finally proceed to download those links into a directory. Having done that will will uncompress the downloaded files which are in a .tgz format

```python
import requests
from bs4 import BeautifulSoup
import re
import os

vox_forge_url = 'http://www.repository.voxforge1.org/downloads/SpeechCorpus/Trunk/Audio/Main/16kHz_16bit/'
page = requests.get(vox_forge_url)
soup = BeautifulSoup(page.content, 'html.parser')
matches = soup.find_all('a', attrs={"href": re.compile("tgz")})
download_links = [vox_forge_url + a['href'] for a in matches]
download_links[:5]
```

```python
downloads_folder = 'audio_samples'
def download_all(matches):
    if not os.path.exists(downloads_folder):
        os.mkdir(downloads_folder)
    total_matches = len(matches)
    for match in matches:
        r = requests.get(vox_forge_url + match['href'] )
        with open(os.path.join(downloads_folder,match['href']),'wb') as f:
            f.write(r.content)
# # WILL DOWNLOAD ALL FILES FOUND        
# download_all(matches)
```

The files are in .tgz format and therefore need to be uncompressed

```python
import tarfile
for file in os.listdir(downloads_folder):
    if file.endswith('.tgz'):
        tar = tarfile.open(os.path.join(downloads_folder, file))
        tar.extractall(downloads_folder)
        tar.close()
```

## Understanding the data and feature extraction
Once the download files are downloaded files are uncompressed the result is a folder containing primarily two subfolders:

1. An **etc** folder

2. A **wav** folder


1. The *etc* folder contains important metadata that pretains to the the individual recordings. Information such as the age, gender and dialect of the individual recording are stored in a README file. Further, the prompts to each recording are usually stored in either a file called prompts-original or PROMPTS. However, at times the files are of concern to use (README and prompts-original) are named differently and at times even missing. I have thus only extracted features where only some variants of these files are present.


2. The *wav* folder contains the actually .wav files of the recordings. The number of recordings by inidividual is not consistent across. Some individuals have 4 recorded samples while others may have 10. Further, the duration of each the recordings are not the same. This is something that will deal with as the feature extracted, more precisely the MFCCs will need to be of the same frames for use to use them in a neural network. Finally, not all of the folders have wav subfolders included in them, as some of the recordings contain FLAC audio files and are subsequntly the folder name is FLAS as opposed to wav. Thus, the features extracted will be limited to the folders that have wav subfolders present


### Extracting the meta data from the readme files
The first step is to extract the age, dialect and most importantly, for the purpose of the task,the gender from the readme files.

```python
from path import Path
#get metadata of audio
def get_meta_data(path_to_readme_directory, readme=''):

    #some readme files are uppercase others are lowercase
    readme_pattern = re.compile(r'readme', flags=re.IGNORECASE)
    readme = None
    #some folders don't have a readme file, so ignore them
    for files in os.listdir(Path(path_to_readme_directory)):
        match = readme_pattern.match(files)    
        if match:
            readme = match.group()
    #some folders don't have a readme(case insensitive) file, so ignore them
    if readme:
        path_to_readme_file = os.path.join(path_to_readme_directory, readme)
        gender, age, dialect = 'N/a', 'N/a', 'N/a'
        for line in open(path_to_readme_file):
            if line.startswith("Gender"):
                gender = re.search("[^: ]*$", line)[0].strip()
            if line.startswith("Age"):
                age = re.search("[^:]*$", line)[0].strip()
            if line.startswith("Pronunciation dialect"):
                dialect = re.search("[^:]*$", line)[0].strip()

        return  gender, age, dialect
    else:
        return None
```

A lot of the data in the readme files is not consistent. Below, we attempt to standardize it. At times gender is capitalize others not for example and sometimes it's not present and thus classified as not_specified. Again for the dialects there are quite a few listed in the data, however for the sake of simplicity only British, American European, Indian and Australian are considered. Other dialects are classified as "Other". Similar logic applies to the age of the speaker.

```python
female_pattern = re.compile('[Ff]emale')
male_pattern = re.compile('[Mm]ale')

american_pattern = re.compile('[Aa]merican')
british_pattern = re.compile('[Bb]ritish')
european_pattern = re.compile('[Ee]uropean')
indian_pattern = re.compile('[Ii]ndian')
australian_pattern = re.compile('[Aa]ustralian')

adult_pattern = re.compile('[Aa]dult')
youth_pattern = re.compile('[Yy]outh')
senior_pattern = re.compile('[Ss]enior')

def standardize_metadata(gender, age_range, dialect):

    #standardize gender
    if female_pattern.search(gender):
        gender = 'Female' 
    elif male_pattern.search(gender):
        gender = 'Male'
    else: 
        gender = 'not_specified'

    #standardize dialect
    if british_pattern.search(dialect):
        dialect = 'British' 
    elif american_pattern.search(dialect)  : 
        dialect = 'American'
    elif european_pattern.search(dialect)  :
        dialect = 'European'
    elif indian_pattern.search(dialect)    : 
        dialect = 'Indian'
    elif australian_pattern.search(dialect): 
        dialect = 'Australian'
    else: dialect = 'Other'

    #age
    if adult_pattern.search(age_range):
        age_range = 'Adult' 
    elif youth_pattern.search(age_range):
        age_range = 'Youth'
    elif senior_pattern.search(age_range):
        age_range = 'Senior'
    else: age_range = 'Unknown'

    return gender, age_range, dialect
```

Next we want to get the actual prompts that the speaker reads from. These are usually located in a file called prompt-original in the /etc/ folder. However, this file isn't always present as is name as "PROMPTS", which we subsequently use to process the prompts.

```python
def get_prompts(path_to_prompt_dir):
    #if prompts-original exists use it else use the PROMPTS file
    
    if os.path.isfile(os.path.join(path_to_prompt_dir,'prompts-original')):
        path_to_prompt_file = os.path.join(path_to_prompt_dir,'prompts-original')
    else: 
        path_to_prompt_file = os.path.join(path_to_prompt_dir,'PROMPTS')
   
    if  os.path.isfile(path_to_prompt_file):                                
        prompt_dict = {}
        for line in open(path_to_prompt_file):
            #ignore empty lines in promptfile
            if len(line.split()) != 0:
            #seperate the wav file name and the prompt - regex is used to match everything after the last forward slash
                sample_file_name = re.findall('([^\/]+$)',re.split("( )",line)[0])[0]
                prompt = "".join(re.split("( )",line)[2:]).strip()
                prompt_dict[sample_file_name] = prompt
        return prompt_dict
    else:
        return None
```

For the purpose of using the data to classify gender based on audio we will need to extract MFCC's from the individual wav files. This is done using the python librosa library. The number of coefficients extracted in this case is 13. Finally, we also calculate the duration of each .wav file

```python
def get_mfcc_and_duration(file_path, n_mfcc=13):
        # Load audio file
        y, sample_rate = librosa.load(file_path)
        duration = librosa.get_duration(filename=file_path)

        # Compute MFCC coefficients
        mfcc = librosa.feature.mfcc(y=y, sr=sample_rate, n_mfcc=n_mfcc)
        return mfcc, duration
```

```python
import librosa
#get the folders of the samples and exclude .tgz files
samples = [x for x in os.listdir(downloads_folder) if os.path.isdir(os.path.join(downloads_folder, x))]
samples_list = []
gender_list =[]
age_list =[]
dialect_list =[]
sample_code_list =[]
text_prompt_list  =[]
duration_list = []
mfcc_list = []
# number_of_folders = len(samples)
# j = 0
for sample in samples:
        
    print(sample)
    sample_folder =  os.path.join(downloads_folder,sample)
    etc_folder = os.path.join(sample_folder, 'etc')
    print(etc_folder)
    sample_wav_folder = os.path.join(sample_folder, 'wav')
    
    #some records don't have a wav folder as the audio format is flac
    if os.path.isdir(sample_wav_folder):
        #make sure the corresonping samples have a readme and prompts file
        if get_meta_data(etc_folder) and get_prompts(etc_folder):
            gender, age, dialect = get_meta_data(etc_folder)
            gender, age, dialect = standardize_metadata(gender, age, dialect)
            prompts_dict = get_prompts(etc_folder)
            for wav_file in os.listdir(sample_wav_folder):
                
                path_to_wav_file = os.path.join(sample_wav_folder,wav_file)
                mfcc, duration = get_mfcc_and_duration(path_to_wav_file)
                
                #some wav files have missing prompts
                if wav_file[:-4] not in prompts_dict.keys():
                    continue
                else:
                    
                    text_prompt_list.append(prompts_dict[wav_file[:-4]])
                    mfcc_list.append(mfcc)
                    duration_list.append(duration)
                    gender_list.append(gender)
                    age_list.append(age)
                    dialect_list.append(dialect)
                    samples_list.append(os.path.join(sample,wav_file))
                
#     j+=1
#     print(str(j) + '/' + str(number_of_folders))
                
```

# 4 Preperaring for deep learning
In this section, we prepare the data for the next step of this task which is classsifying the audio based on the gender. From the visualizations, we first notice that there is a class imbalance between males and females and also that due to odd formats of the readme files, the gender is often not specified. We first proceed to remove those records whose gender isn't specified and then try to balance the data

```python
#subset the data that has the gender specified
gender_data = data[data['gender'] != 'not_specified']
gender_data
```

We also need to encode the gender to 0 and 1 as this is a binary classification task

```python
def gender_encoder(row):
    if row['gender'] =="Male":
        val = 1
    else:
        val = 0 
    return val
gender_data['gender_label'] = gender_data.apply(gender_encoder, axis = 1)
```

```python
gender_data.head()
```

```python
gender_data['gender'].value_counts()
```

For the purposes of machine learning it's always good to have a balanced dataset. As we can see here there are almost five times recordings of males as there are females. We will first subset the male data, shuffle it and select 9396 instances and then concatenate it to the subset of the gender_data for females. This will allow us to have an equal number of recordings for both males and females. In essence we are undersampling the male class so as to achieve an equal ratio of males to females. An alternative approach would be to use Synthetic Minority Oversampling (SMOTE) in order to upsample the minority class which is female in this case. However, I have opted for the first option. The drawback is that we are losing quite a bit of data, however around 18,000 samples should be sufficient.

```python
#subsetting the data to only Males
male_gender_data = gender_data[gender_data['gender'] == 'Male']
#shuffling the data
male_gender_data = male_gender_data.sample(frac=1).reset_index(drop=True)
male_gender_data = male_gender_data[:9396]
male_gender_data
```

```python
#subsetting female data
female_data = gender_data[gender_data['gender'] == 'Female']
#concatenating the two dataframes
balanced_gender_data = pd.concat([male_gender_data, female_data], axis=0)
#ensuring equal number of instances between males and females
balanced_gender_data['gender'].value_counts()
```

Finally we need to ensure that the MFCC's are of equal length. We will add a new column to the dataframe to get the number of frames in each of the the MFCC's. To do this we first find out the maximum number of frames in the MFCC's generated. We then pad the others with zeros to match the length of the one with the most number of frames


```python
balanced_gender_data
```

```python
def get_mfcc_frames(row):
    return row['mfcc'].shape[1]
balanced_gender_data['frames'] = balanced_gender_data.apply(get_mfcc_frames, axis = 1)
balanced_gender_data
```
